Testamos duas vezes o mesmo codigo para ver se o data frame de organizava de forma randomica ou nao. Mas como os valores eram todos iguais fomos tentar de outra maneira, cada vez que reiniciamos o kernel e corremos tudo do inicio a tabela ja estara organizada de outra maneira, por isso dá a entender que é de forma randomica.

Deleçao de genes usando a função .knock_out() tentamos fazer com todos os genes mas o kerner reiniciava e não deixava correr todos os genes do modelo, e os resultados davam sempre os mesmo que deram no single_gene_deletion(). Pode se ver esse código no ficheiro parte1_ex3_a1

Assim aqui vamos fazer deleção de apenas 50 genes e ver se os fluxos de todas as reações associadas se aleteram ou nao

```python
cobra_model4 = cobra_model.copy()
flu = []
for i in range(50):  # Deleção dos 50 primeiros genes, porque só deixava fazer ate 300 genes, mais do que isso o kernel desligava
    cobra_model4.genes[i].knock_out()
    print(cobra_model4.genes[i])
    gene = [cobra_model4.genes[i]]
    solution = cobra_model4.optimize()
    print('***** DEPOIS KO ******')
    print('Objective: ', cobra_model4.objective)
    print('solution.objective_value',solution.objective_value)
    gene.append(solution.objective_value)
    print('fluxo EX_ac_e',solution.fluxes['EX_ac_e'])
    gene.append(solution.fluxes['EX_ac_e'])
    print('fluxo biomassa',solution.fluxes['BIOMASS_Ec_iML1515_core_75p37M'])
    gene.append(solution.fluxes['BIOMASS_Ec_iML1515_core_75p37M'])
    print('fluxo ACtex: ',solution.fluxes['ACtex'])
    print('fluxo ACt4pp: ',solution.fluxes['ACt4pp'])
    print('fluxo ACt2rpp: ',solution.fluxes['ACt2rpp'])
    print('fluxo ACODA: ',solution.fluxes['ACODA'])
    print('fluxo CYSS: ',solution.fluxes['CYSS'])
    print('fluxo UHGADA: ',solution.fluxes['UHGADA'])
    print('fluxo ACKr: ',solution.fluxes['ACKr'])
    flu.append(gene)
    cobra_model4 = cobra_model.copy()
    
```

Ao fazer a deleção dos primeiros 50 genes podemos verificar que o fluxo de acetato extracelular mantem-se igual ao wt nas condições ambientais pedidas)

```python
import pandas as pd

df1 = pd.DataFrame.from_records(flu)
df1.columns = ["KO gene", "solution.objective_value", "fluxo ex_ac_e", "fluxo biomassa"]
df1.sort_values(by= ['fluxo ex_ac_e'])
print(df1)
```

O que não faz sentido aqui é que o solution.objective_value que é o valor da função objetivo (que é maximizar EX_ac_e), em alguns casos não aparece igual ao fluxo de EX_ac_e. No entanto apareceu um aviso no inicio a dizer que algumas soluções eram inviáveis, por isso o que achamos que aconteceu foi que o solver deve ter forcado algumas reaçoes a terem fluxo para o fluxo final do EX_ac_e manter-se no maximo

Estes genes têm indices compreendidos entre 180 e 190, pelo que experimentamos proceder à sua deleção:

```python
cobra_model5 = cobra_model.copy()
flu3 = []
for i in range(180,191):  # Deleção que engloba os genes relacionados com a produção de acetato
    print(cobra_model.genes[i])
    cobra_model5.genes[i].knock_out()
    gene = [cobra_model5.genes[i]]
    solution = cobra_model5.optimize()
    solution.objective_value
    print('***** DEPOIS KO ******')
    print('solution.objective_value',solution.objective_value)
    gene.append(solution.objective_value)
    gene.append(solution.status)
    print('fluxo EX_ac_e',solution.fluxes['EX_ac_e'])
    gene.append(solution.fluxes['EX_ac_e'])
    print('fluxo biomassa',solution.fluxes['BIOMASS_Ec_iML1515_core_75p37M'])
    gene.append(solution.fluxes['BIOMASS_Ec_iML1515_core_75p37M'])
    flu3.append(gene)
    cobra_model5 = cobra_model.copy()
```

```python
import pandas as pd

df2 = pd.DataFrame.from_records(flu3)
df2.columns = ["KO gene", "objective_value", "solution status", "fluxo ex_ac_e", "fluxo biomassa"]
df2.sort_values(by=["objective_value"])
print(df2)
```
